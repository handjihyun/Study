[1] 데이터 준비

```python
import csv
```

```python
FILE = '../DATA/binary-emotion.csv'
```

```python
with open(FILE, mode='r') as f:
    reader = csv.reader(f, delimiter=',')   # delimiter : 구분자

    sentences, emotions = [], []
    for i in reader:
        emotions.append(i[0])
        sentences.append(i[1])
```

[3] 데이터셋 전용 단어사전(VOCA) 생성

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# 데이터셋 사전 관련 상수
NUM_SIZE = 10000
OOV = '<UNK>'
```

```python
tokenizer = Tokenizer(num_words=NUM_SIZE, oov_token=OOV)
```

```python
tokenizer.fit_on_texts(sentences)
```

```python
# ConvertTEXT => NUMBER
tokenizer.texts_to_sequences(sentences)
```
