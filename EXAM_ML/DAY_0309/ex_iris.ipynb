### Feature Scaling and Nomalization
#### ✔ StandardScaler
: 개별 피쳐를 평균이 0이고, 분산이 1인 값으로 변환

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names)
iris_df
```

```python
print(f'feature 들의 평균 값 :\n{iris_df.mean()}\n')
print(f'feature 들의 분산 값 :\n{iris_df.var()}')
```

```python
# Normalize each feature using "StandardScaler"
from sklearn.preprocessing import StandardScaler
```

```python
# Create StandardScaler object
scaler = StandardScaler()

# Convert dataset using StandardScaler / fit() & transform()\
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
```

```python
type(iris_scaled)
```

```python
# Convert "iris_scaled" (Numpy ndarray) to DataFrame
iris_df_scaled = pd.DataFrame(data = iris_scaled, columns = iris_df.columns)
iris_df_scaled
```

```python
print(f'feature 들의 평균 값 :\n{iris_df_scaled.mean()}\n')
print(f'feature 들의 분산 값 :\n{iris_df_scaled.var()}')
```

#### 🔹 학습용 데이터 + 테스트용 데이터 준비

```python
data = iris_df_scaled
data['target'] = iris.target
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :4], data.iloc[:,-1], test_size = 0.2, random_state = 11)
```

#### 🔹학습 모델 생성

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score

y_pre = model.predict(X_test)
accuracy_score(y_test, y_pre)
```

#### ✔ MinMaxScaler
: 데이터값을 0과 1사이의 범위 값으로 변환

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Create MinMaxScaler object
scaler = MinMaxScaler()

# Convert dataset using MinMaxScaler / fit() & transform()
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
```

```python
# Convert "iris_scaled" (Numpy ndarray) to DataFrame
iris_scaled = pd.DataFrame(data = iris_scaled, columns = iris.feature_names)
iris_scaled
```

```python
print(f'feature들의 최솟값 :\n{iris_scaled.min()}\n')
print(f'feature들의 최댓값 :\n{iris_scaled.max()}')
```

```python
data2 = iris_scaled
data2['target'] = iris.target
data2
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data2.iloc[:, :4], data2.iloc[:,-1], test_size = 0.2, random_state = 11)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score

y_pre = model.predict(X_test)
accuracy_score(y_test, y_pre)
```
