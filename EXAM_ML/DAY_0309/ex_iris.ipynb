### Feature Scaling and Nomalization
#### ‚úî StandardScaler
: Í∞úÎ≥Ñ ÌîºÏ≥êÎ•º ÌèâÍ∑†Ïù¥ 0Ïù¥Í≥†, Î∂ÑÏÇ∞Ïù¥ 1Ïù∏ Í∞íÏúºÎ°ú Î≥ÄÌôò

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names)
iris_df
```

```python
print(f'feature Îì§Ïùò ÌèâÍ∑† Í∞í :\n{iris_df.mean()}\n')
print(f'feature Îì§Ïùò Î∂ÑÏÇ∞ Í∞í :\n{iris_df.var()}')
```

```python
# Normalize each feature using "StandardScaler"
from sklearn.preprocessing import StandardScaler
```

```python
# Create StandardScaler object
scaler = StandardScaler()

# Convert dataset using StandardScaler / fit() & transform()\
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
```

```python
type(iris_scaled)
```

```python
# Convert "iris_scaled" (Numpy ndarray) to DataFrame
iris_df_scaled = pd.DataFrame(data = iris_scaled, columns = iris_df.columns)
iris_df_scaled
```

```python
print(f'feature Îì§Ïùò ÌèâÍ∑† Í∞í :\n{iris_df_scaled.mean()}\n')
print(f'feature Îì§Ïùò Î∂ÑÏÇ∞ Í∞í :\n{iris_df_scaled.var()}')
```

#### üîπ ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞ + ÌÖåÏä§Ìä∏Ïö© Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ

```python
data = iris_df_scaled
data['target'] = iris.target
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :4], data.iloc[:,-1], test_size = 0.2, random_state = 11)
```

#### üîπÌïôÏäµ Î™®Îç∏ ÏÉùÏÑ±

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score

y_pre = model.predict(X_test)
accuracy_score(y_test, y_pre)
```

#### ‚úî MinMaxScaler
: Îç∞Ïù¥ÌÑ∞Í∞íÏùÑ 0Í≥º 1ÏÇ¨Ïù¥Ïùò Î≤îÏúÑ Í∞íÏúºÎ°ú Î≥ÄÌôò

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Create MinMaxScaler object
scaler = MinMaxScaler()

# Convert dataset using MinMaxScaler / fit() & transform()
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
```

```python
# Convert "iris_scaled" (Numpy ndarray) to DataFrame
iris_scaled = pd.DataFrame(data = iris_scaled, columns = iris.feature_names)
iris_scaled
```

```python
print(f'featureÎì§Ïùò ÏµúÏÜüÍ∞í :\n{iris_scaled.min()}\n')
print(f'featureÎì§Ïùò ÏµúÎåìÍ∞í :\n{iris_scaled.max()}')
```

```python
data2 = iris_scaled
data2['target'] = iris.target
data2
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data2.iloc[:, :4], data2.iloc[:,-1], test_size = 0.2, random_state = 11)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score

y_pre = model.predict(X_test)
accuracy_score(y_test, y_pre)
```
