```python
import pandas as pd
import numpy as np
```

```python
mart = pd.read_csv('D:\\Study\\EXAM_ML\\ML_project\\walmart.csv', parse_dates=['Date'])
mart.tail()
```

```python
mart['month'] = mart['Date'].dt.month
mart.drop(['Date', 'Holiday_Flag'], axis=1, inplace=True)
```

```python
mart.Temperature = (mart.Temperature - 32) * 5/9
```

```python
mart.Weekly_Sales = np.log(mart.Weekly_Sales)
```

```python
mart
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import *
from sklearn.linear_model import *
from sklearn.metrics import *
```

```python
df, mse, r2score = [], [], []

for i in range(45):
    df.append(mart[mart.Store == i+1])
    df[i].drop('Store', axis=1, inplace=True)
    y = df[i].pop('Weekly_Sales')
    data = df[i]

    X_tr, X_te, y_tr, y_te = train_test_split(data, y, random_state=42, test_size=0.2)
    
    et = ExtraTreesRegressor(max_depth=5)
    et.fit(X_tr, y_tr)
    y_pre = et.predict(X_te)
    mse.append(mean_squared_error(y_te, y_pre))
    r2score.append(r2_score(y_te, y_pre))
```

```python
pd.DataFrame([mse, r2score], index=['mse', 'r2_score']).T.sort_values(by='r2_score', ascending=False)
```

```python
def predict_demand(temp, fuel, cpi, unemploy, month):
    res = et.predict([[temp, fuel, cpi, unemploy, month]])
    
    return str(res[0] * 2.718)
```
