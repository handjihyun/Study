* 데이터 불러오기

```python
# 남북한발전전력량
df = pd.read_excel('../datafiles/dataFiles/남북한발전전력량.xlsx')
df
```

```python
df.dtypes
```

```python
df.iloc[1:5, 0].fillna("남한", inplace=True)
df.iloc[6:9, 0].fillna("북한", inplace=True)
```

```python
df.iloc[4].replace("-", "0", inplace=True)
df.iloc[8].replace("-", "0", inplace=True)
```

```python
df.loc[9] = 0
```

```python
df.iloc[9, [0, 1]] = ['북한', '신재생']
```

```python
df.iloc[:, 2:] = df.iloc[:, 2:].astype("float64")
```

```python
df["평균"] = df.iloc[:, 2:].mean(axis=1)
```
