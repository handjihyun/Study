```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data 로딩 ------------------------------------------------------------
anscombe = sns.load_dataset("anscombe")
anscombe.info()
```

```python
anscombe.head(), anscombe.tail()
```

```python
# 컬럼별로 값의 종류 파악 => 고유값 unique()
anscombe.dataset.unique()
```

```python
anscombe.x.unique()
```

```python
anscombe.y.unique()
```

```python
# '1'인 데이터만 추출
anscombe.dataset == 'I'

oneDF = anscombe[anscombe.dataset == 'I']
```

```python
oneDF.index, oneDF.columns
```

```python
# 데이터 시각화 ================================================
# 1개 데이터 => y축으로 자동 설정
plt.plot(oneDF.y, 'r--')
```

```python
# 2개 데이터 => x축, y축으로 자동 설정
plt.plot(oneDF.x, oneDF.y, 'ro--')
```

```python
# 2개 데이터 => x축, y축으로 자동 설정
plt.plot(oneDF.x, oneDF.y, 'ro')
plt.xlabel("x")
plt.ylabel("y")
plt.title("[anscombe]")
```

```python
# 2개 데이터 => x축, y축으로 자동 설정
plt.plot(oneDF.x, oneDF.y, 'ro', oneDF.x, oneDF.y*0.5, 'g^--')
plt.xlabel("x")
plt.ylabel("y")
plt.title("[anscombe]")
```

```python
# DataFrame.plot() 메서드

oneDF.plot(kind='bar')
```

```python
twoDF = anscombe[anscombe.dataset == 'II']
threeDF = anscombe[anscombe.dataset == 'III']
fourDF = anscombe[anscombe.dataset == 'IV']
```

```python
fig = plt.figure()
```

```python
axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)
```

```python
axes1.plot(oneDF['x'], oneDF['y'])
axes2.plot(twoDF['x'], twoDF['y'])
axes3.plot(threeDF['x'], threeDF['y'])
axes4.plot(fourDF['x'], fourDF['y'])
```
