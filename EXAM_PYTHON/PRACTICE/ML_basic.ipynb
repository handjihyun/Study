#### 딥러닝 기초 개념

- w : weight(가중치)
- b : bias(편향)
- neural network : hidden layer 존재
- feature extraction(특성 추출) : 프로그램이 알아서 특성을 추출함
- node : 숫자 = X(연결된 노드의 값) * w(가중치)
- 오차 : predict - real
- loss function(오차 계산) : sum(오차^2)/n 사용  => 최소화!

#### -------------------------------------------------------------------------
##### 활성함수 (activation function)
* 시그모이드 함수(sigmoid function)
* 탄젠트 함수 (tangent function)
* --> 비선형 예측 (non-linearity)

#### -------------------------------------------------------------------------
##### 최적의 w 찾기
* 첫번 째 w는 랜덤
* 경사하강법 (Gradient Descent) : w - w의 접선의 기울기(w1이 총손실에 미치는 영향)
* 극소점이 아닌 점(local minima)을 찾을 경우가 존재 ==> w - (기울기 * learning rate)
* learning rate(alpha) optimizer : Momentum(가속도 유지), AdaGrad(자주 변하는 w 작게, 자주 변하면 크게), ★Adam

#### -------------------------------------------------------------------------

```python
#딥러닝으로 간단한 수학 문제 풀기
# - 키와 신발사이즈는 관련이 있을까?
키 = [170, 180, 175, 160]
신발 = [260, 270, 265, 255]

## 예시 y = a * 170 + b  -> a, b를 찾는 것이 목표!
```

```python
# 키로 신발사이즈 추론해보기
height = 170
shoe = 260

# shoe = height * a + b

a = tf.Variable(0.1)
b = tf.Variable(0.2)

def lossFunc():
    predict = height * a + b
    return tf.square(260 - predict)

# 경사하강법을 이용해 w값(a, b) 업데이트
# - gradient 자동으로 설정
opt = tf.keras.optimizers.Adam(learning_rate=0.1)

for i in range(300):
    opt.minimize(lossFunc, var_list=[a, b])
#                               -> 경사하강법으로 업데이트 할 w 입력
    print(a.numpy(), b.numpy())
```
